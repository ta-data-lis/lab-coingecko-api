##  Imports

```python
import pandas as pd
import requests
import json
from datetime import datetime
```

# Make a dataframe with Bitcoin and Dogecoin market value for last 30 days vs USD

```python
url_1 = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=USD&days=30'  
resp_1 = requests.get(url_1)

url_2 = 'https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=USD&days=30'
resp_2 = requests.get(url_2)
```

```python
df1 = pd.DataFrame(resp_1.json())
#df1.head()

df2 = pd.DataFrame(resp_2.json())
df2.head()
```

```python
bit = df1['prices'].apply(pd.Series)
bit.columns = ['date','bicoin in USD']
bit.head()

doge = df2['prices'].apply(pd.Series)
doge.columns = ['date','dogecoin in USD']
doge.head()
```

```python
bit['date'] =  pd.to_datetime(bit['date'], unit='ms')
bit.tail()

doge['date'] =  pd.to_datetime(doge['date'], unit='ms')
doge.tail()
```

```python
coin = pd.merge(bit, doge, left_index=True, right_index=True)
coin = coin.drop(['date_y'], axis=1)
coin = coin.rename(columns={'date_x': 'date'})

coin.head()
```

```python
coin['date'] = coin['date'].astype(str)
#print(type(coin['date'][1]))

coin[['date','hour']] = coin['date'].str.split(' ',expand=True)
coin_final = coin.drop(['hour'], axis=1)

#if it gives an error, please re-run the previous cell to redefine 'coin'
```

```python
coin_final.head(50)
```

```python
#Bitcoin and Dogecoin market value for last 30 days (in USD)

coin_bydate = coin_final.groupby('date',axis=0).mean() #groups by date into the mean
coin_bydate.head(30)
```

# Top Ten Coins Report

## List the first 10 coingecko currencies in a DataFrame

```python
url= 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=10&page=1&sparkline=false" -H "accept: application/json'
response = requests.get(url)

top10 = pd.DataFrame(response.json())
top10.head(10)
```

```python
list(top10.columns) 

top10_drop = top10.drop(['image','market_cap',
 'total_volume',
 'high_24h',
 'low_24h',
 'price_change_24h',
 'price_change_percentage_24h',
 'market_cap_change_24h',
 'market_cap_change_percentage_24h',
 'circulating_supply',
 'total_supply',
 'ath',
 'ath_change_percentage',
 'ath_date',
 'roi',
 'last_updated'], axis=1)
```

```python
top10_drop.head(10)
```

```python
top10_coins = top10_drop[0:10]
top10_coins.head(10)
```
