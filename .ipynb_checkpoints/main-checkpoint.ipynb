# Make a dataframe with Bitcoin and Dogecoin market value for last 30 days vs USD

```python
# web scratching for USD/Bitcoin data
response = requests.get('https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=USD&days=30')
ds = pd.DataFrame(response.json())
ds.head(10)
```

```python
# web scratching for USD/dogecoin data
response_dogecoin = requests.get('https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=USD&days=30')
ds_dogecoin = pd.DataFrame(response_dogecoin.json())
ds_dogecoin = ds_dogecoin["prices"].apply(pd.Series)
ds_dogecoin.columns=["date_d","USD/dogecoin"]
ds_dogecoin.head()
```

```python
# show only column "prices"
ds_bitcoin= ds["prices"].apply(pd.Series)
```

```python
# rename columns
ds_bitcoin.columns=["date","USD/Bitcoin"]
ds_bitcoin.head()
```

```python
# transform timestamp to date
from datetime import date


ds_bitcoin["date"] = (pd.to_datetime(ds_bitcoin["date"], unit="ms"))

ds_bitcoin["date"] = (pd.to_datetime(ds_bitcoin["date"]).apply(lambda x: x.date()))
ds_bitcoin.head()
```

```python
# transform timestamp to date
ds_dogecoin["date_d"] = (pd.to_datetime(ds_dogecoin["date_d"], unit="ms"))


ds_dogecoin["date_d"] = (pd.to_datetime(ds_dogecoin["date_d"]).apply(lambda x: x.date()))
ds_dogecoin.head()
```

```python
# Group per day  dogecoin
last_dates_dogecoin = ds_dogecoin.groupby("date_d", as_index=False).agg({"USD/dogecoin":"mean"})
last_dates_dogecoin.head()
```

```python
# Group per day  bitcoin
last_dates = ds_bitcoin.groupby("date", as_index=False).agg({"USD/Bitcoin":"mean"})
last_dates.head()
```

```python
# Concat together Bitcoin and Dogecoin
both_ds = pd.concat([last_dates,last_dates_dogecoin], axis=1)
both_ds.head()
```

```python
# only one date
final_overview = both_ds[["date","USD/Bitcoin","USD/dogecoin"]]
final_overview
```

```python
final_overview.describe()
```

# Top Ten Coins Report

```python
# Top ten dogecoin
ds_topten =final_overview.sort_values(["USD/dogecoin"],ascending=False)
ds_topten.head(10)
```

```python
# Top ten Bitcoin
ds_topten =final_overview.sort_values(["USD/Bitcoin"],ascending=False)
ds_topten.head(10)
```

## List the first 10 coingecko currencies in a DataFrame

```python
# search for the right data on the webside (coin-markets)
coingecko = requests.get('https://api.coingecko.com/api/v3/coins/markets?vs_currency=USD&order=market_cap_desc&per_page=10&page=1&sparkline=false&price_change_percentage=24h')
ds_coingecko = pd.DataFrame(coingecko.json())
ds_coingecko.head(10)
```

```python
ds_coingecko.columns
```

```python
# show only the important columns
ds_coingecko_agg = ds_coingecko[['market_cap_rank',"name","current_price",'total_volume']]
```

```python
# delete index column, just show the ranking of the Top 10
ds_coingecko_agg.set_index("market_cap_rank")
```
