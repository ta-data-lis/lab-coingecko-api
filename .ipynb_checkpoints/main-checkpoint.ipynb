##  Imports

```python
import pandas as pd
import requests
import json
from pandas.io.json import json_normalize
```

# Make a dataframe with Bitcoin and Dogecoin market value for last 30 days vs USD

```python
# Assign page URL to a variable
url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart/range?vs_currency=usd&from=1563753600&to=1566691200'
url1 = 'https://api.coingecko.com/api/v3/coins/dogecoin/market_chart/range?vs_currency=usd&from=1563753600&to=1566691200'
# get method from requests
response = requests.get(url)
response1 = requests.get(url1)
# check if you have the info
print (response)
print (response1)
```

```python
ds_bitcoin = pd.DataFrame(response.json())

ds_dogecoin = pd.DataFrame(response1.json())
```

```python
ds_bitcoin = ds_bitcoin['prices'].apply(pd.Series)
ds_dogecoin = ds_dogecoin['prices'].apply(pd.Series)
```

```python
ds_bitcoin.columns=["date","prices"]
ds_dogecoin.columns=["date","prices"]
```

```python
ds_bitcoin['date'] = (pd.to_datetime(ds_bitcoin['date'], unit='ms'))
ds_dogecoin['date'] = (pd.to_datetime(ds_dogecoin['date'], unit='ms'))
```

```python
ds_bitcoin['date'] = ds_bitcoin['date'].dt.date
ds_dogecoin['date'] = ds_dogecoin['date'].dt.date
```

```python
ds_bitcoin = ds_bitcoin.groupby(['date'], as_index = False).mean()
ds_dogecoin = ds_dogecoin.groupby(['date'], as_index = False).mean()
```

```python
ds_comparison = ds_dogecoin.merge(ds_bitcoin, on = 'date')
ds_comparison.columns=["date","dogecoin","bitcoin"]
```

```python
ds_comparison.head()
```

```python
ds_comparison.sort_values(by='date')
```

# Top Ten Coins Report

```python
url = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=usd&order=market_cap_desc&per_page=100&page=1&sparkline=false'
response = requests.get(url)
print (response)
```

```python
all_coins = pd.DataFrame(response.json())
all_coins
```

```python
all_coins.columns
```

## List the first 10 coingecko currencies in a DataFrame

```python
all_coins = all_coins[['name', 'current_price', 'high_24h', 'market_cap']]
all_coins.head(10)
```
