# Make a dataframe with Bitcoin and Dogecoin market value for last 30 days vs USD

```python
# Dogecoin Dataframe

# Assign page URL to a variable
url = 'https://api.coingecko.com/api/v3/coins/dogecoin/market_chart?vs_currency=USD&days=30'

# get method from requests
response = requests.get(url)#.json()

# check if you have the info
response

type(response)

```

```python
response.text
```

```python
ds = pd.DataFrame(response.json()) 

ds.head() 
```

```python
ds_dropped = ds.drop(['market_caps','total_volumes'], axis = 1)
ds_dropped.head()
```

```python
ds_dogecoin = pd.DataFrame(ds_dropped['prices'].values.tolist(), columns=['date','prices'])
print(ds_dogecoin)
```

```python
ds_dogecoin['date'] = (pd.to_datetime(ds_dogecoin['date'], unit='ms'))
```

```python
ds_dogecoin.head()
```

```python
ds_dogecoin['Dates'] = pd.to_datetime(ds_dogecoin['date']).dt.date
ds_dogecoin['Time'] = pd.to_datetime(ds_dogecoin['date']).dt.time
```

```python
#ds_dogecoin = ds_dogecoin.drop(['date','Time'], axis = 1)
ds_dogecoin.head()
```

```python
ds_dogecoin = ds_dogecoin.groupby('Dates').agg({'prices': 'mean'})
```

```python
ds_dogecoin.shape
ds_dogecoin.head()
```

```python
# Bitcoin Dataframe

# Assign page URL to a variable
url = 'https://api.coingecko.com/api/v3/coins/bitcoin/market_chart?vs_currency=USD&days=30'

# get method from requests
response_bit = requests.get(url)#.json()

# check if you have the info
response_bit

type(response_bit)
```

```python
ds_bit = pd.DataFrame(response_bit.json()) 

ds_bit.head() 

ds_dropped_bit = ds_bit.drop(['market_caps','total_volumes'], axis = 1)
ds_dropped_bit.head()

ds_bitcoin = pd.DataFrame(ds_dropped_bit['prices'].values.tolist(), columns=['date','prices'])

ds_bitcoin['date'] = (pd.to_datetime(ds_bitcoin['date'], unit='ms'))
ds_bitcoin.head()

ds_bitcoin['Dates'] = pd.to_datetime(ds_bitcoin['date']).dt.date
ds_bitcoin['Time'] = pd.to_datetime(ds_bitcoin['date']).dt.time

ds_bitcoin = ds_bitcoin.drop(['date','Time'], axis = 1)
ds_bitcoin.head()

ds_bitcoin = ds_bitcoin.groupby('Dates').agg({'prices': 'mean'})

print(ds_bitcoin.shape)
ds_bitcoin.head()

```

```python
ds_bitcoin = ds_bitcoin.rename(columns={'prices':'Price_Bitcoin_in_USD'})  
ds_dogecoin = ds_dogecoin.rename(columns={'prices':'Price_Dogecoin_in_USD'}) 

ds_dogecoin
```

```python
# Merge Dogecoin & Bitcoin


ds_bit_doge = pd.concat([ds_bitcoin, ds_dogecoin], axis=1)

ds_bit_doge.head()
```

# Top Ten Coins Report

```python
url2 = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=USD&order=market_cap_desc&per_page=100&page=1&sparkline=false'

# get method from requests
response2 = requests.get(url2)#.json()

# check if you have the info
response2

type(response2)
```

```python
ds_Top = pd.DataFrame(response2.json()) 

ds_Top.head()
```

```python
ds_Top = ds_Top[['name','market_cap','market_cap_rank']]
```

```python
ds_Top = ds_Top[:10]
ds_Top
```

## List the first 10 coingecko currencies in a DataFrame

```python
url3 = 'https://api.coingecko.com/api/v3/coins/markets?vs_currency=USD&order=gecko_desc&per_page=100&page=1&sparkline=false'

response3 = requests.get(url3)#.json()

ds_Top_Gecko = pd.DataFrame(response3.json()) 

```

```python
ds_Top_Gecko = ds_Top_Gecko.sort_values(by='market_cap_rank', ascending=True)

ds_Top_Gecko.head(10)
```
